```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

```python
url ='https://www.amazon.in/HP-Pavilion-Anti-Glare-Graphics-14-dv1001TU/product-reviews/B09MH5MLYP/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
page = requests.get(url)
```

```python
page.content
```

```python
soup = bs(page.content,'html.parser')
soup
```

```python
soup.prettify()
```

```python
names = soup.find_all('span',class_='a-profile-name')
names[2:]
```

```python
cust_name =[]
for i in names:
  if i.get_text() not in cust_name:
    cust_name.append(i.get_text())
cust_name
```

```python
title = soup.find_all('a',class_ = 'review-title')
title
```

```python
titles = []
for i in title:
  titles.append(i.getText()[1:-1])
titles

#review_title[:] = [title.strip('\n') for title in review_title]   // can remove \n
```

```python
date = soup.find_all('span',class_ = 'review-date')[2:]
#for i in rev_date:
date
```

```python
rev_date = []
for i in date:
  rev_date.append(i.getText()[21:])
rev_date
```

```python
rating = soup.select('i.review-rating span')[2:]
rating
```

```python
ratings = []
for i in rating:
  ratings.append(i.get_text())
ratings
```

```python
content = soup.select('span.review-text-content')
content
```

```python
rev_content = []
for i in content:
  rev_content.append(i.get_text()[3:-1].strip().replace("\n"," ").replace("\xa0",""))
rev_content
```

```python
# cust_name
# ratings
# titles
# rev_date
# rev_content
```

```python
df = pd.DataFrame()
df['Customer Name'] = cust_name
df['Review Date'] = rev_date
df['Ratings'] = ratings
df['Title of Review'] = titles
df['Review Details'] = rev_content
df
```

```python
df.to_csv('Amazon_HP_Pavillion_reviews.csv',index=False)
```
